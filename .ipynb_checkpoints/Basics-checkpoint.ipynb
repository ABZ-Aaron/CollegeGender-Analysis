# College Gender Gap Basic Analysis

```python
%matplotlib inline
import pandas as pd
import matplotlib.pyplot as plt
```

```python
women_degrees = pd.read_csv('percent-bachelors-degrees-women-usa.csv')
```

```python
women_degrees.head()
```

```python

#Colours for the colourblind
cb_dark_blue = (0/255,107/255,164/255)
cb_orange = (255/255, 128/255, 14/255)

stem_cats = ['Engineering', 
             'Computer Science', 
             'Psychology', 
             'Biology', 
             'Physical Sciences', 
             'Math and Statistics']

fig = plt.figure(figsize=(18, 3))

for sp in range(0,6):
    ax = fig.add_subplot(1,6,sp+1)
    ax.plot(women_degrees['Year'], women_degrees[stem_cats[sp]], c=cb_dark_blue, label='Women', linewidth=3)
    ax.plot(women_degrees['Year'], 100-women_degrees[stem_cats[sp]], c=cb_orange, label='Men', linewidth=3)
    ax.spines["right"].set_visible(False)    
    ax.spines["left"].set_visible(False)
    ax.spines["top"].set_visible(False)    
    ax.spines["bottom"].set_visible(False)
    ax.set_xlim(1968, 2011)
    ax.set_ylim(0,100)
    ax.set_title(stem_cats[sp])
    ax.tick_params(bottom="off", top="off", left="off", right="off")
    
    if sp == 0:
        ax.text(2005, 87, 'Men')
        ax.text(2002, 8, 'Women')
        
    elif sp == 5:
        ax.text(2005, 62, 'Men')
        ax.text(2001, 35, 'Women')
        
plt.show()
```

```python
stem_cats = ['Psychology', 'Biology', 'Math and Statistics', 
             'Physical Sciences', 'Computer Science', 'Engineering']

lib_arts_cats = ['Foreign Languages', 'English', 
                 'Communications and Journalism', 
                 'Art and Performance', 'Social Sciences and History']

other_cats = ['Health Professions', 'Public Administration', 
              'Education', 'Agriculture','Business', 'Architecture']
```

Below demonstrates a way we can generate multiple plots at once using loops. I've also added a legend despite the fact that some of the lines are marked. This is just to show what's possible.

```python
fig = plt.figure(figsize = (15, 20))
fig.suptitle('Gender by Degree', fontsize=24, y = .93)
plt.subplots_adjust(hspace = 0.5)

year = women_degrees['Year']

# Setup charts for STEM.
for index, i in enumerate(range(0, len(stem_cats)*3, 3)):
    ax = fig.add_subplot(6, 3, (i+1))
    
    ax.plot(year, women_degrees[stem_cats[index]], c = cb_dark_blue, linewidth = 3)
    ax.plot(year, 100-women_degrees[stem_cats[index]], c = cb_orange, linewidth = 3)
    ax.tick_params(axis='both', which='both',length=0)
    ax.set_title(stem_cats[index])
    ax.set_yticks([0,100])
    ax.axhline(50, c=(171/255, 171/255, 171/255), alpha=0.3)
    
    # Since x axis labels are the same for all plots, will just remove them from all but
    # the last plot.
    if index != 5:     
        plt.xticks([])
    
    if index == 0:
        ax.text(1975, 72, "Women")
        ax.text(1975, 29, "Men")
        
    if index == 5:
        ax.text(1980, 77, "Men")
        ax.text(1980, 20, "Women")
        
    for value in ax.spines.values():
        value.set_visible(False)
    
# Setup charts for Liberal Arts.
for index, i in enumerate(range(0, len(lib_arts_cats)*3, 3)):
    ax = fig.add_subplot(6, 3, (i+2))
    
    ax.plot(year, women_degrees[lib_arts_cats[index]], c = cb_dark_blue, linewidth = 3)
    ax.plot(year, 100-women_degrees[lib_arts_cats[index]], c = cb_orange, linewidth = 3)
    ax.tick_params(axis='both', which='both',length=0)
    ax.set_title(lib_arts_cats[index])
    ax.set_yticks([0,100])
    ax.axhline(50, c=(171/255, 171/255, 171/255), alpha=0.3)
    
    if index != 4:     
        plt.xticks([])
        
    if index == 0:
        ax.text(1975, 60, "Women")
        ax.text(1975, 40, "Men")
        
    if index == 4:
        ax.text(1980, 65, "Men")
        ax.text(1980, 30, "Women")
        
    for value in ax.spines.values():
        value.set_visible(False)
    
# Setup charts for Other Categories.
for index, i in enumerate(range(0, len(other_cats)*3, 3)):
    ax = fig.add_subplot(6, 3, (i+3))
    
    ax.plot(year, women_degrees[other_cats[index]], c = cb_dark_blue, linewidth = 3, label = "Women")
    ax.plot(year, 100-women_degrees[other_cats[index]], c = cb_orange, linewidth = 3, label = "Men")
    ax.tick_params(axis='both', which='both',length=0, labelbottom='off')
    ax.set_title(other_cats[index])
    ax.set_yticks([0,100])
    ax.axhline(50, c=(171/255, 171/255, 171/255), alpha=0.3)
    
    if index != 5:     
        plt.xticks([])
    
    if index == 0:
        ax.text(1975, 65, "Women")
        ax.text(1975, 30, "Men")
        
    if index == 5:
        ax.text(1980, 77, "Men")
        ax.text(1980, 15, "Women")
        
    for value in ax.spines.values():
        value.set_visible(False)

plt.legend(bbox_to_anchor=(-1.15, .8), loc = 'upper left', prop={'size': 25})
plt.savefig("gender_degrees.png")
plt.show()
```
